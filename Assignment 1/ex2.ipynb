```python
#Lalazov Yunus
import pandas as pd

df = pd.read_csv('earthquakes.csv')

print("Basic Data Info:")
print(df.info())

print("\nFirst 5 rows of the dataset:")
print(df.head())

missing_values = df.isnull().sum()
print("\nMissing values in each column:")
print(missing_values)


df_dropped = df.dropna()
print("\nDataFrame after Dropping Rows:")
print(df_dropped.head())

df_mean_filled = df.fillna(df.select_dtypes(include=['float64', 'int64']).mean())
print("\nDataFrame with Mean Filling:")
print(df_mean_filled.head())

df_median_filled = df.fillna(df.median(numeric_only=True))
print("\nDataFrame with Median Filling:")
print(df_median_filled.head())

df_ffill = df.fillna(method='ffill')
print("\nDataFrame with Forward Filling:")
print(df_ffill.head())

df_bfill = df.fillna(method='bfill')
print("\nDataFrame with Backward Filling:")
print(df_bfill.head())
```
