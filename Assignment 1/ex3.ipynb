```python
#Lalazov Yunus
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler, OneHotEncoder
import numpy as np

df = pd.read_csv('earthquakes.csv')

numerical_columns = ['magnitude', 'depth', 'latitude', 'longitude', 'distanceKM']
min_max_scaler = MinMaxScaler()
df[numerical_columns] = min_max_scaler.fit_transform(df[numerical_columns])

scaler = StandardScaler()
df[numerical_columns] = scaler.fit_transform(df[numerical_columns])

categorical_columns = ['type', 'alert', 'status', 'continent', 'country', 'subnational', 'city']
df_encoded = pd.get_dummies(df, columns=categorical_columns, drop_first=True)

depth_bins = [0, 50, 100, 300, np.inf]
depth_labels = ['Very Shallow', 'Shallow', 'Intermediate', 'Deep']
df['depth_binned'] = pd.cut(df['depth'], bins=depth_bins, labels=depth_labels)

print(df.head())

df_encoded.to_csv('earthquakes_transformed.csv', index=False)
```
